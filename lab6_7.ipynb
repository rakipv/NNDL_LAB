# BPN FOR IRIS DATASET

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = load_iris()
print(" my input parameter\t\n\n",data.data)

# here x are my input (SepalLengthCm ,SepalWidthCm ,PetalLengthCm ,PetalWidthCm
X=data.data
# here y is my taget ( 3 kinds os species - 0:Iris-setosa , 1:Iris-versicolor ,2:Iris-virginica )
y=data.target
```

```python
print(" my target classes \t\n\n",data.target)
```

```python
 #It converts categorical data into dummy or indicator variables
 # here i have choosen y ( target) to be converted ito numerical limit 3
y = pd.get_dummies(y).values
y[:3]
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20)
```

```python
print(y_train)
```

```python
learning_rate = 0.1
iterations = 5000
N = y_train.size
input_size = 4
hidden_size = 2 
output_size = 3  
results = pd.DataFrame(columns=["mse", "accuracy"])
```

```python
# Initialize weights
np.random.seed(10)
# hidden layer
W1 = np.random.normal(scale=0.5, size=(input_size, hidden_size))  
#weight layer
W2 = np.random.normal(scale=0.5, size=(hidden_size , output_size)) 
```

```python
def sigmoid(x):
    
      
    return 1 / (1 + np.exp(-x))

def mean_squared_error(y_pred, y_true):
    return ((y_pred - y_true)**2).sum() / (2*y_pred.size)
    
def accuracy(y_pred, y_true):
    acc = y_pred.argmax(axis=1) == y_true.argmax(axis=1)
    return acc.mean()
```

```python
for itr in range(iterations):    
    
    # feedforward propagation
    # on hidden layer
    Z1 = np.dot(X_train, W1)
    A1 = sigmoid(Z1)

    # on output layer
    Z2 = np.dot(A1, W2)
    A2 = sigmoid(Z2)
    
    
    #  error is calculated herz
    mse = mean_squared_error(A2, y_train)
    acc = accuracy(A2, y_train)
    results=results.append({"mse":mse, "accuracy":acc},ignore_index=True )
    
    # bpn here : go back and updatw weights
    E1 = A2 - y_train
    dW1 = E1 * A2 * (1 - A2)

    E2 = np.dot(dW1, W2.T)
    dW2 = E2 * A1 * (1 - A1)

    
    # weight updates
    W2_update = np.dot(A1.T, dW1) / N
    W1_update = np.dot(X_train.T, dW2) / N

    W2 = W2 - learning_rate * W2_update
    W1 = W1 - learning_rate * W1_update
    print("MSE \n\n\t:" , mse)
    print("ACcuracy \n\n\t :",acc)
    print('weight w1 \n\n\t :',W1)
    print('weight w2 \n\n\t :',W2)
```

BPN IRIS ACCURACY = 0.67

# RBF FOR IRIS DATASET 


IMPORTING NECESSARY PACKAGES

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import math
from sklearn.metrics import accuracy_score
```

LOADING DATASET

```python
dataset = pd.read_csv('/content/iris.csv',sep=None)
dataset.head()
col_to_use = ['sepal.length','sepal.width','petal.length','petal.width']
data = dataset.drop(col_to_use,axis=1)
data.head()
data = data.apply(LabelEncoder().fit_transform)
```

CHOOSEING X AND Y

```python
data_rest = dataset[col_to_use]
data_rest.head()
dataset2 = pd.concat([data_rest,data],axis=1)
dataset2.head()
X= dataset2.drop('variety',axis=1)
y= dataset2['variety']
```

SPLITTING INTO TEST AND TRAIN DATA

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size= 0.33, random_state= 4)
```

SCALE THE DATA USING STANDARD SCALER

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

DETERMINING THE CENTER OF NEURONS USING KMEANS

since neuro should be more than input : we have 4 input so i gave no of neuron( centroid points) to 6

```python
X_train
K_cent= 6
km= KMeans(n_clusters= K_cent, max_iter= 100)
km.fit(X_train)
cent= km.cluster_centers_
```

SETUP THE MATRIX

```python
max=0 
for i in range(K_cent):
	for j in range(K_cent):
		d= np.linalg.norm(cent[i]-cent[j])
		if(d> max):
			max= d
d= max

sigma= d/math.sqrt(2*K_cent)
```

FINDING THE WEIGHT MATRIX TO TRAIN THE NETWORK

```python
shape= X_train.shape
row= shape[0]
column= K_cent
G= np.empty((row,column), dtype= float)

for i in range(row):
  for j in range(column):
    dist= np.linalg.norm(X_train[i]-cent[j])
    G[i][j]= math.exp(-math.pow(dist,2)/math.pow(2*sigma,2))
```

```python
GTG= np.dot(G.T,G)
GTG_inv= np.linalg.inv(GTG)
fac= np.dot(GTG_inv,G.T)
W= np.dot(fac,y_train)
row= X_test.shape[0]
column= K_cent
G_test= np.empty((row,column), dtype= float)
for i in range(row):
	for j in range(column):
		dist= np.linalg.norm(X_test[i]-cent[j])
		G_test[i][j]= math.exp(-math.pow(dist,2)/math.pow(2*sigma,2))
```

ACCURACY

```python
prediction= np.dot(G_test,W)
prediction= 0.5*(np.sign(prediction-0.5)+1)

score= accuracy_score(prediction,y_test)
print(score.mean())
```

RBF IRIS ACCURACY = 0.68
